```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('superconductor.csv')
df
```

```python
for i in df.columns[:-1]:
    print(f"Feature has min {df[i].min():.2f} and max {df[i].max():.2f} ({i})")
```

```python
for i in df.columns[:-1]:
    df[i] = (df[i] - df[i].mean()) / df[i].std()
```

```python
for i in df.columns[:-1]:
    print(f"Feature has min {df[i].min():.2f} and max {df[i].max():.2f} ({i})")
```

```python
df.to_csv('superconductor_processed.csv', sep=',', index = False)
```
